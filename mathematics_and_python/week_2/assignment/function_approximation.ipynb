```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return np.sin(x / 5.0) * np.exp(x / 10.0) + 5.0 * np.exp(-x / 2.0)
```

```python
A_1 = np.array([[1, 1], [1, 15]])
b_1 = np.array([[f(1)], [f(15)]])
```

```python
w1 = np.linalg.solve(A_1, b_1)
```

```python
def f1(w, x):
    return w[0] + w[1] * x
```

```python
x = np.arange(1.0, 15.0, 0.1)
```

```python
plt.plot(x, f(x))
plt.plot(x, f1(w1, x))
plt.show()
```

```python
A_2 = np.array([[1, 1, 1], [1, 15, 15**2], [1, 8, 8**2]])
b_2 = np.array([[f(1)], [f(8)], [f(15)]])
w2 = np.linalg.solve(A_2, b_2)
```

```python
def f2(w, x):
    return w[0] + w[1] * x + w[2] * x**2
```

```python
plt.plot(x, f(x))
plt.plot(x, f2(w2, x))
plt.show()
```

```python
A_3 = np.array([[1, 1, 1, 1], [1, 4, 4**2, 4**3], [1, 10, 10**2, 10**3], [1, 15, 15**2, 15**3]])
b_3 = np.array([[f(1)], [f(4)], [f(10)], [f(15)]])
w3 = np.linalg.solve(A_3, b_3)
```

```python
def f3(w, x):
    return w[0] + w[1] * x + w[2] * x**2 + w[3] * x**3
```

```python
plt.plot(x, f(x))
plt.plot(x, f3(w3, x))
plt.show()
```

```python
result = np.reshape(w3, (1, 4))
print(result)
```
